```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_csv('train.csv')
train_df.head()
```

```python
# Calculate the correlation matrix
correlation_matrix = train_df[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']].corr()

# Plot the correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Toxicity Labels')
plt.show()

# To examine relationships between different toxicity labels
# To identify co-occurrences of labels
# Darker colors (closer to ±1) indicate stronger correlations
# Lighter colors (closer to 0) indicate weaker correlations
# Positive correlations: If insult and obscene are highly correlated, these labels frequently occur together
# Negative Correlation: Uncommon in this context, but would indicate mutual exclusivity
```

```python

# Count the number of comments for each label
label_counts = train_df[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']].sum()

# Plot the class distribution
plt.figure(figsize=(10,6))
sns.barplot(x=label_counts.index, y=label_counts.values, palette='viridis')
plt.title('Class Distribution of Toxicity Labels')
plt.xlabel('Toxicity Type')
plt.ylabel('Number of Comments')
plt.show()

# Indicates the frequency of each label
# Toxic has a much higher count than others e.g threat
# To visualize the distribution of each toxicity label
```

```python
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
```

```python
nltk.download('stopwords')
```

```python
# Get English stopwords
stop_words = set(stopwords.words('english'))

# Combine all toxic and non-toxic comments
toxic_comments = ' '.join(train_df[train_df['toxic'] == 1]['comment_text'].astype(str))
non_toxic_comments = ' '.join(train_df[train_df['toxic'] == 0]['comment_text'].astype(str))

# Generate word cloud for toxic comments
toxic_wc = WordCloud(stopwords=stop_words, background_color='black', colormap='Reds', max_words=200).generate(toxic_comments)
plt.figure(figsize=(10, 6))
plt.imshow(toxic_wc, interpolation='bilinear')
plt.title('Word Cloud for Toxic Comments')
plt.axis('off')
plt.show()

# Generate word cloud for non-toxic comments
non_toxic_wc = WordCloud(stopwords=stop_words, background_color='black', colormap='Greens', max_words=200).generate(non_toxic_comments)
plt.figure(figsize=(10, 6))
plt.imshow(non_toxic_wc, interpolation='bilinear')
plt.title('Word Cloud for Non-Toxic Comments')
plt.axis('off')
plt.show()

# To visualize the most common words used in toxic vs. non-toxic comments
# To understand language patterns associated with each class
# For toxic comments, "know", "fuck", "Nigger" are few of most common toxic comments
# For Non-Toxic comments, "article", "think", "page" are few of most common non-toxic comments
```

```python
# Calculate comment lengths
train_df['comment_length'] = train_df['comment_text'].astype(str).apply(len)

# Plot histogram of comment lengths
plt.figure(figsize=(10,6))
sns.histplot(data=train_df, x='comment_length', hue='toxic', bins=50, palette='viridis', kde=True)
plt.title('Comment Length Distribution (Toxic vs Non-Toxic)')
plt.xlabel('Comment Length')
plt.ylabel('Frequency')
plt.show()

# To compare the length of toxic vs. non-toxic comments
# To explore if comment length is a distinguishing feature
# Peak indicates the most common comment lengths
```
